```python
import pandas as pd
from matplotlib import pyplot as plt

import numpy as np
```

```python
df = pd.read_csv("KS_train_data.csv")
```

```python
df.head(5)
```

```python
df = df.drop(['project_id'], axis=1)
```

```python
df = df.drop_duplicates(subset='project_url', keep="first")
```

```python
df = df.drop(['backers_count', 'converted_pledged_amount', 'name', 'pledged', 'usd_pledged', 
             'project_url', 'reward_url'], axis=1)
```

```python
df.head(5)
```

```python
df['created_at'] = pd.to_datetime(df['created_at'], unit='s')
df['deadline'] = pd.to_datetime(df['deadline'], unit='s')
df['launched_at'] = pd.to_datetime(df['launched_at'], unit='s')
```

```python
df.head(5)
```

```python
df['created_at'] = df['created_at'].dt.normalize()
df['deadline'] = df['deadline'].dt.normalize()
df['launched_at'] = df['launched_at'].dt.normalize()
```

```python
df.head(5)
```

```python
print(df['created_at'].min(), df['created_at'].max())
print(df['launched_at'].min(), df['launched_at'].max())
print(df['deadline'].min(), df['deadline'].max())
```

```python
df['days_to_launch'] = (df['launched_at'] - df['created_at']).dt.days
df['days_to_dealine'] = (df['deadline'] - df['launched_at']).dt.days
```

```python
df.head(5)
```

```python
df['launched_year'] = df['launched_at'].dt.year
```

```python
df = df.drop(['created_at', 'deadline', 'launched_at'], axis=1)
```

```python
df.head(5)
```

```python
df['launched_year'].value_counts()
```

```python
df['is_cat_art'] = df['category'] == 'art'
df['is_cat_music'] = df['category'] == 'music'
df['is_cat_film'] = df['category'] == 'film & video'
df['is_cat_technology'] = df['category'] == 'technology'
df['is_cat_publishing'] = df['category'] == 'publishing'
df['is_cat_food'] = df['category'] == 'food'
df['is_cat_games'] = df['category'] == 'games'
df['is_cat_fashion'] = df['category'] == 'fashion'
df['is_cat_comics'] = df['category'] == 'comics'
df['is_cat_other'] = ~df['category'].isin(['art', 'music', 'film & video', 'technology', 'publishing', 'food', 'games', 'fashion', 'comics'])
```

```python
df.head(5)
```

```python
df = df.drop(['category', 'subcategory'], axis=1)
```

```python
df.head(5)
```

```python
df['is_country_us'] = df['country'] == 'US'
df = df.drop(['country'], axis=1);
```

```python
df.head(5)
```

```python
df['goal_usd'] = df['goal'] * df['fx_rate']
df = df.drop(['goal', 'fx_rate'], axis=1)
```

```python
df.head(5)
```

```python
df = df.drop(['currency'], axis=1)
df['location'] = df['location'].str[-2:]
```

```python
df['location'].value_counts()
```

```python
df['is_loc_ca'] = df['location'] == 'CA'
df['is_loc_ny'] = df['location'] == 'NY'
df['is_loc_uk'] = df['location'] == 'UK'
df['is_loc_tx'] = df['location'] == 'TX'
df['is_loc_cd'] = df['location'] == 'da'
df['is_loc_fl'] = df['location'] == 'FL'
df['is_loc_il'] = df['location'] == 'IL'
df['is_loc_wa'] = df['location'] == 'WA'
df['is_loc_pa'] = df['location'] == 'PA'

df['is_loc_other'] = ~df['location'].isin(['CA', 'NY', 'UK', 'TX', 'da', 'FL', 'IL', 'WA', 'PA'])
```

```python
df = df.drop(['location'], axis=1)
```

```python
df.head(5)
```

```python
df['is_2009'] = df['launched_year'] == 2009
df['is_2010'] = df['launched_year'] == 2010
df['is_2011'] = df['launched_year'] == 2011
df['is_2012'] = df['launched_year'] == 2012
df['is_2013'] = df['launched_year'] == 2013
df['is_2014'] = df['launched_year'] == 2014
df['is_2015'] = df['launched_year'] == 2015
df['is_2016'] = df['launched_year'] == 2016
df['is_2017'] = df['launched_year'] == 2017
df['is_2018'] = df['launched_year'] == 2018
```

```python
df = df.drop(['launched_year'], axis=1)
```

```python
df.head(5)
```

```python
df.to_csv('KS_train_cleaned_1.0.csv', index=False)
```
