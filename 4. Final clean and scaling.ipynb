```python
import pandas as pd
from matplotlib import pyplot as plt

import numpy as np

from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("3_KS_train_w_vader_bayes.csv")
```

```python
df.head(5)
```

```python
df['log_goal_usd'] = np.log10(df['goal_usd'])
df = df.drop(['goal_usd'], axis=1)
```

```python
df.head(5)
```

```python
print(df['days_to_launch'].min(), df['days_to_launch'].max())
print(df['days_to_dealine'].min(), df['days_to_dealine'].max())
```

```python
df['log_days_to_launch'] = np.log(df['days_to_launch'] + 1)
df['log_days_to_dealine'] = np.log(df['days_to_dealine'] + 1)

df = df.drop(['days_to_dealine', 'days_to_launch'], axis=1)
```

```python
df.head(5)
```

```python
print(df['log_days_to_launch'].min(), df['log_days_to_launch'].max())
print(df['log_days_to_dealine'].min(), df['log_days_to_dealine'].max())
```

```python
print(df['log_goal_usd'].min(), df['log_goal_usd'].max())
```

```python
scaler = MinMaxScaler()
scaler.fit(df)
```

```python
df[df.columns] = scaler.fit_transform(df[df.columns]) 
```

```python
df.head(5)
```

```python
df.to_csv('KS_train_final.csv', index=False)
```
