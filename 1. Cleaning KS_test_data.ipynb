```python
import pandas as pd
from matplotlib import pyplot as plt

import numpy as np
```

```python
df['launched_year'] = df['launched_at'].dt.year
```

```python
df.head(5)
```

```python
df['is_country_us'] = df['country'] == 'US'
df = df.drop(['country'], axis=1);
```

```python
df['is_2009'] = df['launched_year'] == 2009
df['is_2010'] = df['launched_year'] == 2010
df['is_2011'] = df['launched_year'] == 2011
df['is_2012'] = df['launched_year'] == 2012
df['is_2013'] = df['launched_year'] == 2013
df['is_2014'] = df['launched_year'] == 2014
df['is_2015'] = df['launched_year'] == 2015
df['is_2016'] = df['launched_year'] == 2016
df['is_2017'] = df['launched_year'] == 2017
df['is_2018'] = df['launched_year'] == 2018
```

```python
df = df.drop(['launched_year'], axis=1)
```

```python
df.head(5)
```

```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
analyser = SentimentIntensityAnalyzer()
```

```python
df['blurb'].fillna(" ", inplace=True)

for index, value in df['blurb'].items():
    s = analyser.polarity_scores(value)
    df.at[index, 'sentiment_com'] = s['compound']
    
    df.at[index, 'sentiment_pos'] = (df.at[index, 'sentiment_com'] >= 0.05)
    df.at[index, 'sentiment_neu'] = (df.at[index, 'sentiment_com'] > -0.05 and df.at[index, 'sentiment_com'] < 0.05)
    df.at[index, 'sentiment_neg'] = (df.at[index, 'sentiment_com'] <= -0.05)
    
df['sentiment_pos'] = df['sentiment_pos'].astype('bool')
df['sentiment_neu'] = df['sentiment_neu'].astype('bool')
df['sentiment_neg'] = df['sentiment_neg'].astype('bool')
```

```python
df = df.drop(['sentiment_com'], axis=1)
```
