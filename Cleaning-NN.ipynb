```python
import pandas as pd
from matplotlib import pyplot as plt

import numpy as np
```

```python
df = pd.read_csv("KS_train_data_clean_1.0.csv")
```

```python
df.head(5)
```

```python
df = df.drop(df.columns[0], axis=1)
```

```python
df.head(5)
```

```python
df['is_cat_art'] = df['category'] == 'art'
df['is_cat_music'] = df['category'] == 'music'
df['is_cat_film'] = df['category'] == 'film & video'
df['is_cat_technology'] = df['category'] == 'technology'
df['is_cat_publishing'] = df['category'] == 'publishing'
df['is_cat_food'] = df['category'] == 'food'
df['is_cat_games'] = df['category'] == 'games'
df['is_cat_fashion'] = df['category'] == 'fashion'
df['is_cat_comics'] = df['category'] == 'comics'
df['is_cat_other'] = ~df['category'].isin(['art', 'music', 'film & video', 'technology', 'publishing', 'food', 'games', 'fashion', 'comics'])
```

```python
df.head(5)
```

```python
df = df.drop(['category', 'subcategory'], axis=1)
```

```python
df.head(5)
```

```python
df = df.drop(['blurb'], axis=1)
```

```python
df.head(5)
```

```python
df['is_country_us'] = df['country'] == 'US'
df = df.drop(['country'], axis=1);
```

```python
df.head(5)
```

```python
df['goal_usd'] = df['goal'] * df['fx_rate']
df = df.drop(['goal', 'fx_rate'], axis=1)
```

```python
df.head(5)
```

```python
df = df.drop(['currency'], axis=1)
```

```python
df.head(5)
```

```python
df['location'] = df['location'].str[-2:]
```

```python
df.head(5)
```

```python
df['location'].describe()
```

```python
df['location'].value_counts()
```

```python
df['is_loc_ca'] = df['location'] == 'CA'
df['is_loc_ny'] = df['location'] == 'NY'
df['is_loc_uk'] = df['location'] == 'UK'
df['is_loc_tx'] = df['location'] == 'TX'
df['is_loc_cd'] = df['location'] == 'da'
df['is_loc_fl'] = df['location'] == 'FL'
df['is_loc_il'] = df['location'] == 'IL'
df['is_loc_wa'] = df['location'] == 'WA'
df['is_loc_pa'] = df['location'] == 'PA'

df['is_loc_other'] = ~df['location'].isin(['CA', 'NY', 'UK', 'TX', 'da', 'FL', 'IL', 'WA', 'PA'])
```

```python
df.head(5)
```

```python
df = df.drop(['location'], axis=1)
```

```python
df.head(5)
```

```python
df.to_csv('KS_final_train.csv', index=)
```
