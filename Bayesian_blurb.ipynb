```python
import pandas as pd
from matplotlib import pyplot as plt

import numpy as np

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv("KS_train_data_clean_1.0.csv")
```

```python
df.head(5)
```

```python
df['blurb'].fillna(" ", inplace=True)
```

```python
for index, value in df['blurb'].items():
    s = "".join([ c if c.isalpha() else " " for c in value ])
    s = s.lower()
    
    df.at[index, 'clean_blurb'] = s
```

```python
df.head(5)
```

```python
X = df.clean_blurb
y = df.funded
```

```python
X.fillna(" ", inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
vectorizer = CountVectorizer()
counts = vectorizer.fit_transform(X.values)
```

```python
classifier = MultinomialNB(alpha=1)
targets = y.values
classifier.fit(counts, targets)
```

```python
example_count = vectorizer.transform(X.values)
predictions = classifier.predict(example_count)
```

```python
predictions
```

```python
print(classification_report(y, predictions))
```
