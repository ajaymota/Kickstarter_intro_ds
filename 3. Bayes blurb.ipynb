```python
import pandas as pd
from matplotlib import pyplot as plt

import numpy as np

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv("KS_train_w_vader.csv")
```

```python
df.head(5)
```

```python
df['blurb'].fillna(" ", inplace=True)
```

```python
for index, value in df['blurb'].items():
    s = "".join([ c if c.isalpha() else " " for c in value ])
    s = s.lower()
    
    df.at[index, 'clean_blurb'] = s
```

```python
df.head(5)
```

```python
X = df.clean_blurb
y = df.funded
```

```python
X.fillna(" ", inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10)
```

```python
vectorizer = CountVectorizer()
counts = vectorizer.fit_transform(X_train.values)
```

```python
classifier = MultinomialNB(alpha=1)
targets = y_train.values
classifier.fit(counts, targets)
```

```python
example_count = vectorizer.transform(X_test.values)
predictions = classifier.predict(example_count)
```

```python
print(classification_report(y_test, predictions))
```

```python
example_count = vectorizer.transform(X.values)
x = classifier.predict_proba(example_count)
# print(x)
i = 0

for row in x:
    df.at[i, 'blurb_bayes'] = row[1]
    i+=1
```

```python
df['funded'].corr(df['blurb_bayes'])
```

```python
df = df.drop(['clean_blurb', 'blurb'], axis=1)
```
