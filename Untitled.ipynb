```python
from pandas import read_csv
from matplotlib import pyplot as plt

import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import r2_score, mean_squared_error, explained_variance_score, max_error

from keras.models import Sequential
from keras.models import model_from_json
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
from keras.optimizers import Adam
from keras.callbacks import CSVLogger
```

```python
# Plot original dataset to visualize it
df = read_csv('airline-passengers.csv', header=0, index_col=0)
df.plot()
plt.show()
```

```python
# Normalize returns to make predictions more effective
df['Passengers'] = np.log(df['Passengers'])
df.plot()
plt.show()
```

```python
# Split dataset into training and test data
X = df
X_train, X_test = train_test_split(X, test_size=0.25, shuffle=False)

print(len(X_train), len(X_test))
```

```python
# Create moving values for Training Set
y_train = X_train.shift(-1, axis=0)
y_train = y_train[:-1]

X_train = X_train[:-1]

# Create moving values for Test Set
y_test = X_test.shift(-1, axis=0)
y_test = y_test[:-1]

X_test = X_test[:-1]
```

```python
print(X_train.iloc[:4])
print(y_train.iloc[:4])
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
input_data = scaler.fit_transform(X_train.values)
predi_data = scaler.fit_transform(y_train.values)

# Build 3D featureset for LSTM
features_set = np.reshape(input_data, (input_data.shape[0], input_data.shape[1], 1))
```

```python
model = Sequential()
model.add(LSTM(units=4, input_shape=(features_set.shape[1], 1)))

model.add(Dense(units=1))

opt = Adam(lr=0.001)
model.compile(optimizer=opt, loss='mean_squared_error')
print("Model created successfully!")
```

```python
csv_logger = CSVLogger('training.log')
model.fit(features_set, predi_data, epochs=100, batch_size=8, callbacks=[csv_logger])
```

```python
learn = read_csv('training.log', header=0, index_col=0)
learn.plot()
plt.show()
```

```python
test_inputs = X_test.values
test_inputs = test_inputs.reshape(-1, 1)

test_inputs = scaler.fit_transform(test_inputs)

test_inputs = np.reshape(test_inputs, (test_inputs.shape[0], test_inputs.shape[1], 1))

predictions = model.predict(test_inputs)

predictions = scaler.inverse_transform(predictions)
```

```python
del X_test['Passengers']
X_test['Predicted Passengers'] = predictions

ax = y_test.plot()
X_test.plot(ax=ax)
plt.show()
```

```python
r2 = r2_score(X_test, predictions)
mse = mean_squared_error(X_test, predictions)
mae = max_error(X_test, predictions)

print("R^2 = ", round(r2*100, 3), "%")
print("Mean-Squared Error = ", round(mse*100, 3), "%")
print("Maximum Error = ", round(mae*100, 3), "%")
```
